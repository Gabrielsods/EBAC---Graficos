```python

import pandas as pd
import requests
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# %matplotlib notebook

# 1) seu código aqui
df = pd.read_csv('SINASC_RO_2019.csv')
df.drop_duplicates().shape

```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO,ESTCIVMAE '.

```python
######## RESPOSTA - INCOMPLETA ##########

plt.rc('figure', figsize=(10, 8))
fig, axes = plt.subplots(3, 2)

sns.countplot(ax = axes[0, 0], x='APGAR1', data=df)
sns.countplot(ax = axes[0, 1], x='APGAR5', data=df)
sns.countplot(ax = axes[1, 0], x='SEXO', data=df)

y = sns.countplot(ax = axes[1, 1], x='GESTACAO', data=df)
y.set_xticklabels(y.get_xticklabels(),rotation = 30)

sns.countplot(ax = axes[2, 0], x='ESTCIVMAE', data=df)

plt.subplots_adjust(wspace=0.3, hspace=1)
```

```python
###### CORREÇÃO INSTRUTOR #############

variaveis = ['APGAR1','APGAR5','SEXO','GESTACAO','ESTCIVMAE']

for var in variaveis:
    fig = plt.figure(); ax = fig.add_subplot(1,1,1)
    sns.countplot(data = df, x = var)
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.rc('figure', figsize=(10, 8))
fig, axes = plt.subplots(2, 2)

sns.countplot(ax = axes[0, 0], x='SEXO', data=df)
y = sns.countplot(ax = axes[0, 1], x='GESTACAO', data=df)
sns.countplot(ax = axes[1, 0], x='ESTCIVMAE', data=df)
sns.countplot(ax = axes[1, 1], x='CONSULTAS', data=df)

y.set_xticklabels(y.get_xticklabels(),rotation = 90)
plt.subplots_adjust(wspace=0.3, hspace=0.8)
```
